## **CATEGORIZE DATA**

```python
# Define a function to map residency requirements to broader categories
def map_residency_requirement(residency):
    if 'NYC Residency' in residency or 'New York City residency' in residency:
        return 'NYC Residency Required'
    elif 'NY State residency' in residency or 'New York State Residency' in residency:
        return 'NY State Residency Required'
    elif 'NYCHA' in residency:
        return 'NYCHA Residency'
    elif 'No residency requirement' in residency or 'not required' in residency:
        return 'No Residency Requirement'
    else:
        return 'Other Residency Requirement'

# Create a new column 'Residency Category' by applying the mapping function
new_df['Residency'] = new_df['Residency Requirement'].apply(map_residency_requirement)

# Display the unique values in the new column
print(new_df['Residency'].unique())
```

```python
# Define a function to map original job categories to broader categories
def map_category(original_category):
    if 'Constituent Services & Community Programs' in original_category:
        return 'Community Programs'
    elif 'Technology, Data & Innovation' in original_category:
        return 'Technology and Data'
    elif 'Social Services' in original_category:
        return 'Social Services'
    elif 'Building Operations' in original_category:
        return 'Building Operations'
    elif 'Finance, Accounting, & Procurement' in original_category:
        return 'Finance and Procurement'
    elif 'Engineering, Architecture, & Planning' in original_category:
        return 'Engineering and Construction'
    elif 'Legal Affairs' in original_category:
        return 'Legal Affairs'
    elif 'Administration & Human Resources' in original_category:
        return 'Administration and Human Resources'
    elif 'Policy, Research & Analysis' in original_category:
        return 'Policy, Research & Analysis'
    elif 'Intergovernmental' in original_category:
        return 'Intergovernmental Affairs'
    elif 'Public Safety, Inspections, & Enforcement' in original_category:
        return 'Public Safety and Enforcement'
    elif 'Health' in original_category:
        return 'Health'
    else:
        return 'Multi Category'

# Apply mapping to create a new column for broader categories
new_df['Domain'] = new_df['Job Category'].apply(map_category)

# Display the unique values in the new column
print(new_df['Domain'].unique())
```

# **PREDICTION MODEL**

### **Lets start with PCA and RandomForestRegressor**




```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

# Combine the encoded 'Level' column with other features
X = pd.concat([new_df[['# Of Positions', 'Salary Range To', 'Salary Range From']], pd.get_dummies(new_df['Level'], prefix='Level')], axis=1)
y = new_df['Avg Salary']  # Target variable

# Handle missing values if any
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Define the pipeline with PCA and RandomForestRegressor
pipeline = Pipeline([
    ('pca', PCA(n_components=0.95)),  # Keep 95% of variance
    ('rf_regressor', RandomForestRegressor(random_state=42))
])

# Define the parameter grid to search
param_grid = {
    'rf_regressor__n_estimators': [50, 100, 200],
    'rf_regressor__max_depth': [None, 10, 20],
    'rf_regressor__min_samples_split': [2, 5, 10],
    'rf_regressor__min_samples_leaf': [1, 2, 4]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_
print("Best Parameters:", best_params)

# Get the best model
best_model = grid_search.best_estimator_

# Predict on the test set
y_pred = best_model.predict(X_test)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Calculate R-squared score
r2 = r2_score(y_test, y_pred)
print("R-squared Score:", r2)

# Calculate accuracy
threshold = 5000  # Define a threshold for accuracy evaluation
accurate_predictions = abs(y_pred - y_test) <= threshold
accuracy = sum(accurate_predictions) / len(y_test)
print("Accuracy (within {} threshold): {:.2f}%".format(threshold, accuracy * 100))
```
